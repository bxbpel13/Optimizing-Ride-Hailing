```python
import pandas as pd
import pyarrow.parquet as pq

# List of file paths for your 6 Parquet files
file_paths = ["/Users/bhavikpatel/Desktop/poject MSDA/Data 228/Project/yt/yellow_tripdata_2023-07.parquet",
             "/Users/bhavikpatel/Desktop/poject MSDA/Data 228/Project/yt/yellow_tripdata_2023-08.parquet",
             "/Users/bhavikpatel/Desktop/poject MSDA/Data 228/Project/yt/yellow_tripdata_2023-09.parquet"]

# Initialize an empty DataFrame
trips = pd.DataFrame()

# Loop through each file and read it into a DataFrame, then append to the combined DataFrame
for file_path in file_paths:
    # Read Parquet file into a DataFrame
    parquet_file = pq.read_table(file_path)
    df = parquet_file.to_pandas()
    
    trips = pd.concat([trips, df], ignore_index=True)


# Display the combined DataFrame
print(trips)
```
